```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.boxplot(data = iris , palette = "pastel")
plt.title("Boxplot of Iris dataset")
plt.show()
```

```python
sepal_data = iris['sepal_width']
sns.boxplot(data = sepal_data , palette = "pastel")
plt.title("Sepal Width of Iris dataset")
plt.show()
```

```python
sq1 = sepal_data.quantile(0.25)
sq3 = sepal_data.quantile(0.75)
print("Quantile 25%",sq1)
print("Quantile 75%",sq3)
sepal_iqr = sq3 - sq1
sepal_lower_bound = sq1 - (1.5 * sepal_iqr)
sepal_upper_bound = sq3 + (1.5 * sepal_iqr)
print("Lower Bound",sepal_lower_bound)
print("Upper Bound",sepal_upper_bound)
filtered_sepal_data = sepal_data[(sepal_data >= sepal_lower_bound) & (sepal_data <= sepal_upper_bound)]
filtered_sepal_data
```

```python
sns.boxplot(data=filtered_sepal_data, palette="pastel")
plt.title("Sepal Width of Iris dataset")
plt.show()
```

```python
iris["sepal_width"] = filtered_sepal_data
sns.boxplot(data = iris["sepal_width"] , palette = "pastel")
plt.title("Sepal Width of Iris dataset")
plt.show()
```
